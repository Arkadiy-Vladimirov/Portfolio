***Goal:*** $\varepsilon \sim \mathrm{Exp}(\lambda)$  
***Implementation:*** $\varepsilon = -\frac{1}{\lambda} \ln \eta,\;$ where 
$\eta \sim \mathrm{Uni}(0,1)$

```python
def gen_exponential(l, n):
    """Generates random sample [X1, X2,.. X_n] where Xi ~ Exp(l)"""
    return [(-1.0/l)*math.log(np.random.random()) for i in range(n)]

sns.displot({"Value": gen_exponential(2,100000)}, x = "Value", stat="probability",
                                            kde=True, kde_kws={"bw_adjust": 1})
```

***Goal:*** $\pi \sim \mathrm{Pois}(\lambda)$  
***Implementation:*** $\pi = \max\limits_{\mathbb{N}\cup\{0\}}(n \mid S_n = \sum_{i=1}^{n} \varepsilon_i \le 1)$

```python
def gen_poisson_e(l,n):
    """Generates random sample [X1, X2,.. X_n] where Xi ~ Pois(l)"""
    out = []
    for i in range(n):
        sm = 0
        num = 0
        while sm <= 1:
            sm += gen_exponential(l,1)[0]
            num += 1
        num -= 1
        out.append(num)
    return out

sns.displot({"Value": gen_poisson_e(4,10000)}, x="Value", stat="probability",
                                                                discrete = True)
```

***Given:*** $\beta \sim \mathrm{Bin}(n,p)$  
***Goal:***  $\pi \sim \mathrm{Pois}(\lambda)$  
***Implementation:*** $\pi(\lambda) \sim \lim_{
    \substack{n \rightarrow \infty \\ p \rightarrow 0 \\ np \rightarrow \lambda}}
    \beta(n,p)$

```python
def gen_fbinomial(n,p, size): #fast version
    """Generates random sample [X1, X2,.. X_{size}] where Xi ~ Bi(n,p)"""
    return np.sum((np.random.rand(size,n) < p), axis=1)

def gen_poisson_b(l,n,infty = 1000):
    """Generates random sample [X1, X2,.. X_{n}] where Xi ~ Pois(l)"""
    p = l/infty
    return gen_fbinomial(infty,p,n)

sns.displot({"Value": gen_poisson_b(4,10000)}, x="Value", stat="probability", discrete = True)
    
```

***Goal:*** perform $\chi^2$ test for goodness of fit $\pi \sim \mathrm{Pois}(\lambda)$  

```python
n,l,alpha = 1000, 4, 0.05
N = math.ceil(math.log2(n))

d = np.arange(N) #d[N-1] --- smth other than 0..N-1
p = sps.poisson.pmf(d,l)
p[N-1] = 1 - np.sum(p[0:-1]) #p[N-1] --- prob of evrth other than 0..N-1
x = gen_poisson_b(l,n)
nu = [np.size(x[x == i]) for i in range(N)]
nu[N-1] = n - np.sum(nu[:-1])

Xn2 = np.sum(((nu-n*p)**2) / (n*p))
sps.chi2.cdf(Xn2,N-1) < 1-alpha
```

***Goal:*** $(X,Y) \sim \mathcal{N}(0,1)$  
***Implemenation:***  
$X = -\sqrt{-2\ln\eta_1}\cos(2\pi\eta_2)$,  
$Y = -\sqrt{-2\ln\eta_1}\sin(2\pi\eta_2)$,  
where $\eta_1, \eta_2 \sim \mathrm{U}[0,1]$.


```python
def gen_normal(n):
    """Generates random sample [X1, X2,.. X_{n}] where Xi ~ N(0,1)"""
    eta = np.random.rand(2,n)
    r = -np.sqrt(-2*np.log(eta[0,:]))
    phi = 2*np.pi*eta[1,:]
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return np.array([x,y])

sample = gen_normal(1000000)
sns.displot({"Value": sample[0,:]}, x = "Value", stat="density",
                                            kde=True, kde_kws={"bw_adjust": 1})                                            
```

***Goal:*** perform t- and F- tests to check equality of X and Y distributions

```python
n = m = 1000
alpha = 0.05
XY = gen_normal(n)
barX = 1/n * np.sum(XY[0,:])
barY = 1/m * np.sum(XY[1,:])
S12 = 1/(n-1) * np.sum((XY[0,:]-barX)**2)
S22 = 1/(m-1) * np.sum((XY[1,:]-barY)**2)
Fstat = S12/S22
Sstat = np.sqrt(n*m/(n+m)) * (barX-barY) * (n+m-2) / ((n-1)*S12 + (m-1)*S22)

alpha < sps.f.cdf(Fstat,n-1,m-1) < 1-alpha,
alpha < sps.t.cdf(Sstat,n+m-2) < 1-alpha
```
